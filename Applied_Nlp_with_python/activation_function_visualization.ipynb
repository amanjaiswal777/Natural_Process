```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_sigmoid():
    
    def sigmoid(x):
        return (1/float(1+np.exp(-x)))
    
    def derive_sigmoid(x):
        return (sigmoid(x)* (1 - sigmoid(x)))
    ''' np.r_ Translates slice objects to concatenation along the first axis.

This is a simple way to build up arrays quickly. There are two use cases.

If the index expression contains comma separated arrays, then stack them along their first axis.
If the index expression contains slice notation or scalars then create a 1-D array with a range indicated by the slice notation.'''
    sigmoid_array, derive_array = [], []
    x = np.r_[-6:6, 6]
    
    for i in x:
        sigmoid_array.append(sigmoid(i))
        derive_array.append(derive_sigmoid(i))
        
    sigmoid, = plt.plot(sigmoid_array, label=['f(x)'])
    derive, = plt.plot(derive_array, label=['df/dx'])
    plt.title('Sigmoid Activation Function')
    plt.xlabel('X Value')
    plt.ylabel('Activation function output')
    plt.xticks(range(0, len(x)), x)
    plt.legend(handles=[sigmoid, derive])
```

```python
plot_sigmoid()
```

```python
def plot_relu():
    
    def relu(x):
        return max(0,x)
    
    def derive_relu(x):
        if x < 0: return 0
        elif x > 0:return 1
        else: return 0.5
        
    relu_array, derive_array = [], []
    x = np.r_[-6:6, 6]
    for i in x:
        relu_array.append(relu(i))
        derive_array.append(derive_relu(i))
        
    relu, = plt.plot(relu_array, label=['relu'])
    derive, = plt.plot(derive_array, label=['df/dx'])
    plt.title('ReLU Activation Function')
    plt.xlabel('X Value')
    plt.ylabel('Activation Function Output')
    plt.xticks(range(0, len(x)), x)
    plt.legend(handles=[relu, derive])
```

```python
plot_relu()
```

```python
def plot_tanh():
    
    def tanh(x):
        return ((np.exp(x)-np.exp(-x))/float(np.exp(x) + np.exp(-x)))
    
    def derive_tanh(x):
        return (1 - tanh(x)**2)
    
    tanh_array, derive_array = [], []
    
    x = np.r_[-6:6, 6]
    for i in x:
        tanh_array.append(tanh(i))
        derive_array.append(derive_tanh(i))
        
        
    tanh, = plt.plot(tanh_array, label=['f(x)'])
    derive, = plt.plot(derive_array, label=['df/dx'])
    plt.title('Tanh Activation Function')
    plt.xlabel('X Value')
    plt.ylabel('Activation Function Output')
    plt.xticks(range(0, len(x)), x)
    plt.legend(handles=[tanh, derive])
    
```

```python
plot_tanh()
```
