```python
import jellyfish
correct_example = ('Narendra Modi', 'Narendra Modi')
damodardas_example = ('Narendra Modi', 'Narendra D. Modi')
modi_type_example = ('Narendra Modi', 'Narender Modi')
gujrat_type_example = ('Gujarat', 'Gujrat')
```

```python
examples = [correct_example, damodardas_example, modi_type_example, gujrat_type_example]
```

```python
def calculate_distance(function, examples=examples):
    for i in examples:
        print(f'{i}: {function(*i)}')
```

```python
calculate_distance(jellyfish.levenshtein_distance)
```

```python
calculate_distance(jellyfish.damerau_levenshtein_distance)
```

```python
calculate_distance(jellyfish.hamming_distance)
```

```python
calculate_distance(jellyfish.jaro_distance)
```

```python
calculate_distance(jellyfish.jaro_winkler)
```

```python
from flashtext.keyword import KeywordProcessor
keyword_processor = KeywordProcessor()
keyword_processor.add_keyword('TBC', 'USTECH')
keyword_processor.add_keyword('Noida', 'Delhi')
keywords_found = keyword_processor.extract_keywords('I love working with the USTECH in Delhi')
```

```python
keywords_found
```

```python
replaced_sentence = keyword_processor.replace_keywords('I love working with Machine learning algorithm')
```
